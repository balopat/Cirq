```python
import datetime 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure

xs = []
ys = []
colors = []

with open("prs.csv") as f: 
    for l in f.read().split("\n"):
        if not ";" in l:
            continue
        rec = eval(l.split(";")[1])        
        xs.append(rec[1])
        ys.append(rec[4].total_seconds()/60/60/24)
        colors.append("red" if rec[5] == "open" else "green")
        
figure(num=None, figsize=(20, 10), dpi=80)        
plt.scatter(xs,ys,color=colors)
plt.show()
```
